```python
from pathlib import Path
from snowsar.utils.nisar_utils import (
    extract_gunw_layers_to_geotiff_batch,
)
```

```python
gunw_dir = Path("/Volumes/Fortress_L3/SDC/NISAR_Tropo/NISAR_CA_T135_F20-21/products/")       # <- edit
pattern = "*.h5"                    # or "NISAR_L2_PR_GUNW_*.h5"
out_dir = Path("/Users/havazli/Desktop/outtifs")    # <- edit

# Choose which frequency/pol you want
frequency = "A"
pol = "HH"
```

```python
layers = ["unwrappedPhase", "coherenceMagnitude", "ionospherePhaseScreen", "connectedComponents",
          "incidenceAngle", "localIncidenceAngle", "totalTroposphere"]  # <- edit
outputs = extract_gunw_layers_to_geotiff_batch(
    gunw_dir=gunw_dir,
    pattern=pattern,
    out_dir=out_dir,
    frequency="A",
    pol="HH",
    layers=layers,
    warp=True,
    dst_epsg=4326,          # or any EPSG you want
    dst_res=None,
    resampling="nearest",
    overwrite=False,
)
```

```python
# quick check
first_file = next(iter(outputs))
outputs[first_file]
```

```python
import rasterio as rio
from rasterio.plot import show
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize = (8, 8))

with rio.open(outputs[first_file]['unwrappedPhase']) as src:
    data = src.read(1, masked=True)
    im = show(data, ax=ax, transform=src.transform)
    # fig.colorbar(im.get_images()[0], ax=ax, fraction=0.036, pad=0.04)
```
