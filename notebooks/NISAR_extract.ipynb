```python
from pathlib import Path
from snowsar.utils.nisar_utils import (
    extract_gunw_layers_to_geotiff_batch,
)
```

```python
gunw_dir = Path("/Users/havazli/Desktop/SnowWaterEquivalent/NISAR_data_sample/")       # <- edit
pattern = "*.h5"                    # or "NISAR_L2_PR_GUNW_*.h5"
out_dir = Path("/Users/havazli/Desktop/outtifs_test")    # <- edit

# Choose which frequency/pol you want
frequency = "A"
pol = "HH"
```

```python
layers = ["unwrappedPhase", "coherenceMagnitude", "ionospherePhaseScreen", "connectedComponents",
          "incidenceAngle", "localIncidenceAngle", "totalTroposphere", 'losUnitVectorX', 'losUnitVectorY',
          'elevationAngle', 'slantRangeSolidEarthTidesPhase']  # <- edit as needed
outputs = extract_gunw_layers_to_geotiff_batch(
    gunw_dir=gunw_dir,
    pattern=pattern,
    out_dir=out_dir,
    frequency="A",
    pol="HH",
    layers=layers,
    warp=True,
    dst_epsg=4326,          # or any EPSG you want
    dst_res=None,
    resampling="nearest",
    overwrite=False,
)
```

```python
# quick check
first_file = next(iter(outputs))
outputs[first_file]
```

```python
import rasterio as rio
from rasterio.plot import show
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize = (8, 8))

with rio.open(outputs[first_file]['localIncidenceAngle_interp']) as src:
    data = src.read(1, masked=True)
    im = show(data, ax=ax, transform=src.transform, cmap='jet')
```
